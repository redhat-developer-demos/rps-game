```python
import requests
import json
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
from remote_infer import process_image

# Define KServe inference endpoint
KSERVE_URL = "https://yolov11-kcn-rock-paper-scissors.apps.xxx/v2/models/yolov11/infer"

# Define the local image path
HOME = os.getcwd()
destination_dir = os.path.join(HOME, "rock-paper-scissors-14")  # Root directory
image_name = "egohands-public-1622748405009_png_jpg.rf.94b2a32daad24b1e1963b47feceac216.jpg"
image_path = os.path.join(destination_dir, "test", "images", image_name)
```

```python
# Perform inference
img = process_image(image_path, KSERVE_URL)

# Convert BGR to RGB for visualization
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the result
plt.figure(figsize=(8, 6))
plt.axis("off")
plt.imshow(img)
plt.show()
```
